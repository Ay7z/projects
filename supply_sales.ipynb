```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\AYAZ\OneDrive\Desktop\datasets\supply_chain_data.csv')
```

```python
df.columns
```

```python
revenue = df.groupby('Product type')['Revenue generated'].sum().reset_index()
revenue.columns = ['Product type','Total revenue']
revenue
```

```python
products_sold =df.groupby('Product type')['Number of products sold'].sum().reset_index()
products_sold.columns = ['Product type', 'Products sold' ]
products_sold = products_sold.sort_values(by='Products sold', ascending=False)
products_sold
```

```python
df['Defect rates']
```

```python
cost_cuts = df.groupby('Product type')[['Shipping costs','Manufacturing costs','Costs']].sum()
cost_cuts
```

```python
merged = pd.merge(revenue, cost_cuts, on='Product type')
```

```python
merged['profit'] = merged['Total revenue']-merged['Shipping costs']-merged['Manufacturing costs']-merged['Costs']
```

```python
merged[['Product type', 'Total revenue', 'profit']]
```

Key insight:
Skincare wins on both revenue and total profit. But cosmetics has proportionally lower costs, meaning per rupee earned, cosmetics is more efficient. Haircare underperforms despite decent unit sales.

Business takeaway:
Never rank product lines by revenue alone. A high-selling product with bloated costs can be less valuable than a lower-selling one with lean operations.
